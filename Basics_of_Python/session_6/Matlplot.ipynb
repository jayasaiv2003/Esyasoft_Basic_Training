```python
import matplotlib.pyplot as plt
```

```python
x=[0,2,4,6,8]
y=[0,4,16,36,64]
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,marker='o',label="Data Points")
ax.set_title("Basic cpmponents of Matplotlib Figure")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
```
